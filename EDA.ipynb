```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
#Lectura de archivo gracias a la librería pandas
archivo = pd.read_csv("alzheimers_disease_data.csv")
#archivo = archivo.drop("DoctorInCharge",axis=1)
```

## Se muestran los valores únicos presentes

```python
#Visualización de valores únicos presentes en el dataset.
cols = archivo.columns.tolist()
for col in cols:
  print(f"{col}: {archivo[col].unique()}")
```

## Se eliminan columnas innecesarias

```python
#Se elimina la columna DoctorInCharge, dado que solo presenta un valor que no entrega información
archivo = archivo.drop("DoctorInCharge",axis=1)
archivo = archivo.drop ("PatientID",axis=1)
archivo.head()
```

```python
#Comprobar que no existen valores duplicados
print(archivo.duplicated().sum())
```

## Generación de matriz de correlación

```python
#Creación de una matriz de correlación de Pearson.
correlacion = archivo.select_dtypes(include="number")
sns.heatmap(archivo.corr(method="pearson"), annot=False, cmap='coolwarm', annot_kws={"size": 6})
plt.show()
```

## Generación de gráficos, histogramas

```python
#Visualización de la distribución de diagnósticos positivos y negativos en el dataset en base a la edad.
plt.figure(figsize=(10, 6))
ax =sns.histplot(data=archivo, x='Age', hue='Diagnosis', multiple='stack', bins=12, palette='Set2')
plt.title('Distribución del Diagnóstico en cuanto a la edad')
plt.xlabel('Edad')
plt.ylabel('Cantidad')
handles, labels = ax.get_legend_handles_labels()
ax.legend(title='Diagnóstico', labels=['Positivo', 'Negativo'])
plt.show()
```

```python
#Visualización de la distribución de diagnósticos positivos y negativos en el dataset en base al índice de masa corporal.
plt.figure(figsize=(10, 6))
ax = sns.histplot(data=archivo, x='BMI', hue='Diagnosis', multiple='stack', bins=12, palette='Set2')
plt.title('Distribución del Diagnóstico en cuanto al índice de masa corporal')
plt.xlabel('Índice de masa corporal')
plt.ylabel('Cantidad')
handles, labels = ax.get_legend_handles_labels()
ax.legend(title='Diagnóstico', labels=['Positivo', 'Negativo'])
plt.show()
```

## Generación de gráficos, countplot

```python
#Visualización de cantidad de problemas de comportamiento según el diagnóstico.
labels = ["No", "Sí"]
ticks = range(len(labels))
ax = sns.countplot(data=archivo, x='Diagnosis', hue='BehavioralProblems', palette="Set2")
plt.xticks(ticks=ticks, labels=labels)
plt.xlabel("Diagnóstico")
plt.ylabel("Cantidad")
plt.title('Cantidad de problemas de comportamiento dado el diagnóstico')
handles, _ = ax.get_legend_handles_labels()
ax.legend(handles=handles, labels=["No presenta", "Presenta"], title="Problemas de Comportamiento")
plt.show()
```

```python
#Visualización de cantidad de historial familiar de Alzheimer según el diagnóstico.
labels = ["No", "Sí"]
ticks = range(len(labels))
ax = sns.countplot(data=archivo, x='Diagnosis', hue='FamilyHistoryAlzheimers', palette="Set2")
plt.xticks(ticks=ticks, labels=labels)
plt.xlabel("Diagnóstico")
plt.ylabel("Cantidad")
plt.title('Cantidad de historial familiar de alzheimer dado el diagnóstico')
handles, _ = ax.get_legend_handles_labels()
ax.legend(handles=handles, labels=["No presenta", "Presenta"], title="Historial familiar de alzheimer")

plt.show()
```

```python
#Visualización de cantidad de personas que presentan falta de memoria según el diagnóstico.
labels = ["No", "Sí"]
ticks = range(len(labels))
ax = sns.countplot(data=archivo, x='Diagnosis', hue='Forgetfulness', palette="Set2")
plt.xticks(ticks=ticks, labels=labels)
plt.xlabel("Diagnóstico")
plt.ylabel("Cantidad")
plt.title('Cantidad de personas que presentan falta de memoria dado el diagnóstico')
handles, _ = ax.get_legend_handles_labels()
ax.legend(handles=handles, labels=["No presenta", "Presenta"], title="Falta de memoria")
plt.show()
```

```python
#Visualización de cantidad de personas con desorientación según el diagnóstico.
labels = ["No", "Sí"]
ticks = range(len(labels))
ax = sns.countplot(data=archivo, x='Diagnosis', hue='Disorientation', palette="Set2")
plt.xticks(ticks=ticks, labels=labels)
plt.xlabel("Diagnóstico")
plt.ylabel("Cantidad")
plt.title('Cantidad de personas que presentan desorientación dado el diagnóstico')
handles, _ = ax.get_legend_handles_labels()
ax.legend(handles=handles, labels=["No presenta", "Presenta"], title="Desorientación")
plt.show()
```
